# 1. Data type changes(astype)

```python
import pandas as pd

data = {
    'Product_ID': [101, 102, 103, 104, 105],
    'Product_Name': ['iPhone', 'MacBook Pro', 'iPad', 'Apple Watch', 'AirPods'],
    'Release_Year': ['2020', '2019', '2018', '2021', '2019'],
    'Price ($)': [999, 1299, 329, 399, 159],
    'Units_Sold_Millions': ['75', '20', '40', '25', '30']
}

apple_df = pd.DataFrame(data)
apple_df
```

```python
apple_df.dtypes
```

#### How would you convert the 'Release_Year' column from string type to integer type? 

```python
apple_df.astype({'Release_Year': 'int64'}).dtypes
```

#### Can you use astype to change the data type of the 'Price ($)' column to a float?

```python
apple_df.astype({'Price ($)': 'float64'}).dtypes
```

#### How can you change the 'Units_Sold_Millions' column from string to integer data type?


```python
apple_df.astype({'Units_Sold_Millions': 'int64'}).dtypes
```

#### If you wanted the 'Product_ID' column to be of string type for joining with another dataset, how would you use astype to achieve this?

```python
apple_df.astype({'Product_ID': 'str'}).dtypes
```

#### Can you transform the 'Units_Sold_Millions' column into a float type and ensure that the column represents the actual number of units (i.e., not in millions) using a single line of code?

```python
apple_df.loc[:, ['Units_Sold_Millions']].astype('float64').apply(lambda x: x * 1000000)
```

#### How would you convert all string columns in the DataFrame to uppercase while changing their data type to a category?

```python
apple_df.astype('category').dtypes
```

```python
apple_df.select_dtypes(include=['object']).applymap(lambda x: x.upper() if isinstance(x, str) else x).astype('category')
```

# 5. Concat, extend and merge

```python
data = {
    'Shop_ID' : [1, 2, 3, 4, 2, 5, 6],
    'Shop_Name': ['Starbucks', 'Blue Bottle', 'Verve Coffee', 'Stumptown', 'Blue Bottle', 'Gregorys Coffee', 'Cafe Grumpy'],
    'Location': ['Hollywood', 'San Diego', 'Hollywood', 'San Diego', 'New York', 'New York', 'New York'],
    'Rating': [4.5, 4.8, 4.4, 4.3, 4.6, 4.2, 4.7],
    'Revenue': [50000, 30000, 25000, 28000, 52000, 20000, 24000]
}
coffee_shops = pd.DataFrame(data)
coffee_shops
```

```python
suppliers_data = {
'Shop_ID': [1, 2, 3, 4, 5, 6, 7],
'Shop_Name': ['Starbucks', 'Blue Bottle', 'Verve Coffee', 'Stumptown', 'Gregorys Coffee', 'Blue Bottle', 'Gregorys Coffee'],
'Supplier_Name': ['Beans R Us', 'Premium Beans', 'South Coffee Suppliers', 'San Diego Beans', 'NY Fresh Beans', 'Mexico tea', 'Chayyos'],
'Delivery_Days': [7, 5, 7, 4, 6,9, 10]
}

suppliers_df = pd.DataFrame(suppliers_data)
coffee_shops
```

### Basic Merge:
   - How would you merge coffee_shops with suppliers_df based on a common column named Shop_ID?

```python
coffee_shops.merge(suppliers_df, on='Shop_ID')
```

### Left Merge:
  - If you wanted to retain all records from coffee_shops and only the matching records from suppliers_df based on the Shop_ID column, which type of merge would you use?

```python
suppliers_df.loc[5] = [7, 'Chaayos', 'Tata Coffee', 9]
coffee_shops
```

```python
suppliers_df
```

```python
coffee_shops.merge(suppliers_df, on='Shop_ID', how='left')
```

### Right Merge:
  - Assuming that suppliers_df has some Shop_ID entries that don't exist in coffee_shops, how would you merge such that you retain all supplier records and only matching coffee shop records?


```python
coffee_shops.merge(suppliers_df, on='Shop_ID', how='right')
```

### Outer Merge:
   - If you wanted to retain all records from both coffee_shops and suppliers_df, including those without matching Shop_ID, which merge strategy would you employ?

```python
coffee_shops.merge(suppliers_df, on='Shop_ID', how='outer')
```
